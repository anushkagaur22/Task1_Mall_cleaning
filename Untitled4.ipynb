```python
import pandas as pd
import numpy as np
```

```python
#Read the upolad file
df = pd.read_csv("/content/Mall_Customers.csv")
```

```python
df.columns = [c.strip().lower().replace(" ","_").replace("-","_")for c in df.columns]
```

```python
df.drop_duplicates(inplace=True)
```

```python
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype == "object":
            df[col].fillna("unknown", inplace=True)
        else:
            df[col].fillna(df[col].median(), inplace=True)
```

```python
if "gender" in df.columns:
  df["gender"] = df["gender"].str.strip().str.lower().replace({
      "male":"Male","female":"Female"
  })
```

```python
num_cols = ["age", "annual_income_(k$)", "spending_score_(1_100)"]

for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

# Outlier removal (IQR)
for c in num_cols:
    Q1, Q3 = df[c].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    low, high = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    df = df[(df[c] >= low) & (df[c] <= high)]

print("Cleaned shape:", df.shape)
df.head()
```

```python

for c in num_cols:
    Q1 = df[c].quantile(0.25)
    Q3 = df[c].quantile(0.75)
    IQR = Q3 - Q1
    low = Q1 - 1.5 * IQR
    high = Q3 + 1.5 * IQR
    before_rows = df.shape[0]
    df = df[(df[c] >= low) & (df[c] <= high)]
    after_rows = df.shape[0]
    print(f"Removed {before_rows - after_rows} outliers in {c}")

print("\nFinal shape after cleaning:", df.shape)

print("\nMissing values after cleaning:\n", df.isnull().sum())

df.head()
```

```python
df.to_csv("mall_customers_cleaned.csv", index=False)

```
